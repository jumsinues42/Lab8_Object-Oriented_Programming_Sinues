# 1. Abstract Base Classes 

```python
from abc import ABC, abstractmethod 
 
class Animal(ABC): 
    @abstractmethod 
    def sound(self): 
        pass 
```

# 2. Data Classes for Cleaner Code 

```python
from dataclasses import dataclass 
 
@dataclass 
class Student: 
    name: str 
    age: int 
    course: str
```

# 3. Custom Exception Handling 

```python
class InvalidAgeError(Exception): 
    pass 
 
def set_age(age): 
    if age < 0: 
        raise InvalidAgeError("Age cannot be negative")
```

# 4. Use Property Decorators for Controlled Access 

```python
class Temperature: 
    def __init__(self, celsius): 
        self._celsius = celsius 
 
    @property 
    def fahrenheit(self): 
        return (self._celsius * 9/5) + 32
```

# 5. Favor Composition Over Inheritance for Flexibility 

```python
class Engine: 
    def start(self): 
        print("Engine started") 
 
class Car: 
    def __init__(self): 
        self.engine = Engine() 
 
    def drive(self): 
        self.engine.start() 
        print("Car is moving")
```
