```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class ShoppingCart:
    def __init__(self):
        self.items = []
        self.total_price = 0

    def add_product(self, product):
        if product.stock > 0:
            self.items.append(product)
            product.stock -= 1
            self.total_price += product.price
            print(f"Added {product.name} to cart.")
        else:
            print(f"Error: {product.name} is out of stock.")

    def remove_product(self, product_name):
        for item in self.items:
            if item.name.lower() == product_name.lower():
                self.items.remove(item)
                item.stock += 1
                self.total_price -= item.price
                print(f"Removed {item.name} from cart.")
                return
        print("Error: Item not in cart.")

    def display_cart(self):
        print("\n--- Your Cart ---")
        for item in self.items:
            print(f"- {item.name}: ${item.price}")
        print(f"Total: ${self.total_price:.2f}")

# Predefined Inventory
inventory = [
    Product("Laptop", 1200, 5),
    Product("Mouse", 25, 10),
    Product("Keyboard", 50, 2)
]

# Example Usage
cart = ShoppingCart()
cart.add_product(inventory[0]) # Laptop
cart.add_product(inventory[2]) # Keyboard
cart.display_cart()
cart.remove_product("Laptop")
cart.display_cart()
```
