```python
class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_payroll(self):
        overtime_hours = max(0, self.hours_worked - 160)
        base_hours = self.hours_worked - overtime_hours
        
        base_salary = base_hours * self.hourly_rate
        overtime_pay = overtime_hours * (self.hourly_rate * 1.5)
        total_salary = base_salary + overtime_pay
        
        return base_salary, overtime_pay, total_salary

# Example Usage
employees = []
n = int(input("Enter number of employees: "))
if n <= 0:
    print("Error: Must enter at least one employee.")
else:
    for i in range(n):
        name = input(f"Name of Employee {i+1}: ")
        rate = float(input("Hourly rate: "))
        hours = float(input("Total hours worked: "))
        
        if rate < 0 or hours < 0:
            print("Invalid input: Negative values rejected.")
            continue
            
        employees.append(Employee(name, rate, hours))

    print("\n--- Payroll Summary ---")
    for emp in employees:
        base, ovt, total = emp.calculate_payroll()
        print(f"Name: {emp.name} | Hours: {emp.hours_worked} | Base: ${base:.2f} | Overtime: ${ovt:.2f} | Total: ${total:.2f}")
```
