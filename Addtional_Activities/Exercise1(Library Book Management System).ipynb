```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))
        print(f"Book '{title}' added successfully.")

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    print(f"You have successfully borrowed '{book.title}'.")
                    return
                else:
                    print(f"Error: '{book.title}' is already borrowed.")
                    return
        print("Error: Book not found.")

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.is_available = True
                print(f"You have successfully returned '{book.title}'.")
                return
        print("Error: Book not found.")

    def display_books(self):
        print("\n--- Library Inventory ---")
        for book in self.books:
            status = "Available" if book.is_available else "Borrowed"
            print(f"Title: {book.title} | Author: {book.author} | Status: {status}")

# Example Usage
my_library = Library()
num_books = int(input("Enter number of books to add: "))
for _ in range(num_books):
    t = input("Title: ")
    a = input("Author: ")
    my_library.add_book(t, a)

my_library.display_books()
my_library.borrow_book(input("\nEnter title to borrow: "))
my_library.display_books()
```
