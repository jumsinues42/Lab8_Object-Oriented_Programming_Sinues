# 1. Creating a Class and Object in Python

```python
#Step 1: Define a class named Car
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def display_info(self):
        return f"{self.year} {self.brand} {self.model}"

# Step 2: Create an object of the Car class
my_car = Car ("Toyota", "Corolla", 2022)

# Step 3: Print the car details using the method
print(my_car.display_info())
```

# 2.  Implementing Encapsulation in Python 

```python
# Step 1: Define a class with private attributes
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self._balance = balance # Private attribute
    def deposit (self, amount):
        self._balance += amount
        return f"New balance: \${self._balance}"
    def get_balance(self):
        return f"Balance: \${self._balance}"

# Step 2: Create an object and test encapsulation
account = BankAccount("Alice", 5000)

# Step 3: Deposit money and display the updated balance
print(account.deposit (1500))
print(account.get_balance())
```

## Try it part for number 2:

```python
class StudentRecord:
    def __init__(self):
        self.__grades = []  # Private attribute

    def add_grade(self, grade):
        self.__grades.append(grade)

    def get_all_grades(self):
        return self.__grades

# Test encapsulation
record = StudentRecord()
record.add_grade(95)
record.add_grade(88)
print(f"Grades List: {record.get_all_grades()}")
```

# 3. Applying Inheritance in Python 

```python
#Step 1: Define a base class
class Person:
    def __init__(self, name):
        self.name = name
    def introduce(self):
        return f"Hi, I'm {self.name}."

#Step 2: Define a subclass that inherits from Person
class Student (Person):
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a student."

# Step 3: Create objects and call methods
person = Person ("Allen")
student = Student("Mark")
print(person.introduce())
print(student.introduce())
```

# 4. Utilizing Polymorphism in Python 

```python
# Step 1: Define a base class
class Shape:
    def area(self):
        return "Calculating area..."

# Step 2: Define subclasses with their implementations
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle (Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

# Step 3: Create objects and call their methods
circle = Circle (5)
rectangle = Rectangle(4, 6)
print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
```

# 6. Working with Class and Static Methods 

```python
# Step 1: Define a class with a static method
class MathOperations:
    @staticmethod
    def square(num):
        return num * num

# Step 2: Get user input
user_input = int(input("Enter a number to square: "))

# Step 3: Call the static method with user input
result = MathOperations.square(user_input)

#Step 4: Print the result
print("Square of", user_input, ":", result)
```

## Try it part for number 6:

```python
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

# User interaction
c_input = float(input("Enter temperature in Celsius: "))
f_result = TemperatureConverter.celsius_to_fahrenheit(c_input)
print(f"{c_input}°C is equal to {f_result}°F")
```

# 7. Operator Overloading in Python 

```python
# Step 1: Define a class with overloaded operators
class BankAccount:
    def __init__(self, balance): 
        self.balance = balance

    def __add__(self, other):
        return BankAccount(self.balance + other.balance)

# Step 2: Create account objects and add their balances
account1 = BankAccount(5000)
account2 = BankAccount(3000)
total_balance = account1 + account2 # Combines the balances

# Step 3: Print the result
print(f"Total Balance: P{total_balance.balance}")
```

## Try it part for number 7:

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Demonstrate vector math
v1 = Vector(5, 10)
v2 = Vector(2, 3)
print(f"Addition: {v1 + v2}")
print(f"Subtraction: {v1 - v2}")
```

# 8. Abstract Classes and Methods in Python 

```python
from abc import ABC, abstractmethod

# Step 1: Define an abstract class
class Appliance (ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Define a subclass implementing the abstract method
class WashingMachine (Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create an object of the subclass and call the method
wm = WashingMachine()
print(wm.turn_on())
```

# 9. Implementing Method Resolution Order (MRO) 

```python
#Step 1: Define multiple classes
class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B,C):
    pass # Inherits show() based on MRO

# Step 2: Create an object of class D and call the method
obj = D()
print(obj.show()) # Output depends on MRO
```

## Try it part for number 9:

```python
class A:
    def show(self):
        print("Message from Class A")

class B(A):
    def show(self):
        print("Message from Class B")

class C(A):
    def show(self):
        print("Message from Class C")

class D(B, C):
    def show(self):
        print("Message from Class D")

# Explore MRO
print("Method Resolution Order for Class D:")
print(D.mro())

obj_d = D()
obj_d.show()
```
